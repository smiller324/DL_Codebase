```python
import json
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# rho = 5000
data5000 = pd.read_json("case_793_5000.json").T.reset_index()
data5000.sort_values("index", inplace = True)

# rho = 1
data0001 = pd.read_json("case_793_0001.json").T.reset_index()
data0001.sort_values("index", inplace = True)

def format_data(df):
    df = df.values.tolist()
    index = []
    value = []
    
    for i in df:
        index.append(i[0])
        value.append(i[1])
        
    return (index, value)

# rho = 5000
(iters, gen)    = format_data(data5000[["index", "gen"]])
(iters, dual)   = format_data(data5000[["index", "dual"]])
(iters, primal) = format_data(data5000[["index", "primal"]])

# rho = 1
(iters, gen0001)    = format_data(data0001[["index", "gen"]])
(iters, dual0001)   = format_data(data0001[["index", "dual"]])
(iters, primal0001) = format_data(data0001[["index", "primal"]])
```

```python
plt.plot(iters, gen, label = "rho = 5000")
# plt.plot(iters2, gen1000, label = "rho = 1e9")
plt.plot(iters, gen0001, label = "rho = 1")

# plt.plot(iters, [28.5] * len(iters), label = "Real Power Load")

plt.legend(['rho = 5000', "rho = 1"])
plt.xlabel('Iterations')
plt.ylabel('Real Power Generated (MWh)')
```

```python
plt.plot(iters, primal, label = "rho = 5000")
plt.plot(iters, primal0001, label = "rho = 1")

plt.legend(['rho = 5000', "rho = 1"])
plt.xlabel('Iterations')
plt.ylabel('Max Primal Infeasibility')
```

```python
plt.plot(iters, dual, label = "rho = 5000")
plt.plot(iters, dual0001, label = "rho = 1")
# plt.plot(iters, dual0001, label = "rho = 1")

# plt.plot(iters3, dual3, label = "rho = 1000")

plt.legend(['rho = 5000', "rho = 1"])
plt.xlabel('Iterations')
plt.ylabel('Max Dual Infeasibility')
```
